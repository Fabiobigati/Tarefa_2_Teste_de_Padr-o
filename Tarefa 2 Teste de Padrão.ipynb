```python
def mae (y, ypred):
    yDiff = ypred - y.reshape(-1,1)
    yAbs = abs(yDiff)
    errosum = sum(yAbs)
    return errosum/len(yAbs)
print(mae(y,ypred))
```

```python
def mse (y, ypred):
    yDiff = ypred - y.reshape(-1,1)
    yDiffSquare = yDiff**2
    errosum = sum(yDiffSquare)
    return errosum/len(ypred)
print(mse(y,ypred))
```

```python
from math import sqrt
def rmse (y, ypred):
    yDiff = ypred - y.reshape(-1,1)
    yDiffSquare = yDiff**2
    errosum = sum(yDiffSquare)
    mse = errosum/len(ypred)
    return sqrt(mse)
print(rmse(y,ypred))
```

```python
from sklearn.metrics import mean_squared_error

def rmse (y, ypred):
    mse = mean_squared_error(y, ypred)
    return sqrt(mse)

print(rmse(y,ypred))
```

```python
class regresso_linear():
    def __init__(self):
        self.a = 0
        self.b = 0
    def predict(self, X, y=None):
        ypred = self.a * X + self.b
        return ypred
```

```python
modelo = regresso_linear()

ypred = modelo.predict (X12)

plt.scatter(X12, y)
plt.plot(X12, ypred, 'r')
print(mean_squared_error(y,ypred))
```

```python
class regresso_linear1():
    def __init__(self):
        self.a = 0
        self.b = 0
    def predict(self, X, y=None):
        ypred = self.a * X + self.b
        return ypred        
    def fit (self, X, y):
        for _ in range(500):
            ypred = self.predict (X)
            erros = ypred - y.reshape(-1,1)
            ajusteB = sum(erros)/len(erros)
            print("Ajuste B: ",ajusteB)
            self.b = self.b - ajusteB*0.1
            ajusteA = sum(erros[i]*X[i] for i in range (X.shape[0]))/sum(X)
            self.a = self.a - ajusteA*0.1
            print("Ajuste A",ajusteA)
            ypred = self.predict(X)
            print("Error: ",mean_squared_error(y,ypred))
            

modelo1 = regresso_linear1()
modelo1.fit(X12, y)
ypred = modelo1.predict(X12)
plt.scatter(X12, y)
plt.plot(X12, ypred)
```

```python
import pandas as pd
import numpy as np
np.random.seed(0)
from matplotlib import pyplot as plt

dados = pd.DataFrame()
dados['x'] = np.linspace(-10,10,100)
dados['y'] = 5 + 3*dados['x'] + np.random.normal(0,3,100)

plt.scatter(dados['x'], dados['y'])
plt.axhline(y=0, linewidth=2, color = 'k')
plt.axvline(x=0, linewidth=2, color = 'k')
plt.show()
```

```python
from IPython import display
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
%matplotlib notebook


def L(y, y_hat):
    return ((y-y_hat) ** 2).sum()


b_hat, w_hat = np.linspace(-10,20,40), np.linspace(0,6,40)


loss = np.array([L(dados['y'], i + j * dados['x']) for i in b_hat for j in w_hat]).reshape(40,40)
b_hat, w_hat = np.meshgrid(b_hat, w_hat)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_zticks([])
ax.set_xlabel('$\hat{b}$')
ax.set_ylabel('$\hat{w}$')
ax.set_zlabel('Custo', rotation=90)
surf = ax.plot_surface(b_hat, w_hat, loss,
                       rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0)
plt.show()
```

```python
class linear_regr(object):

    def __init__(self, learning_rate=0.0001, training_iters=1000, show_learning=False):
        self.learning_rate = learning_rate
        self.training_iters = training_iters
        self.show_learning = show_learning

        
    def fit(self, X_train, y_train, plot=False):
        
        
        if len(X_train.values.shape) < 2:
            X = X_train.values.reshape(-1,1)
        X = np.insert(X, 0, 1, 1)
        
        
        if self.show_learning:
            assert X.shape[1] <= 2 
            self.w1_loss = [L(y_train, 5 + i * X_train) for i in np.linspace(0,6,20)]
            self.b_loss = [L(y_train, i + 3 * X_train) for i in np.linspace(-10,20,20)]
        
        
        self.w_hat = np.random.normal(0,5, size = X[0].shape)
        
        loss = [] 
        for _ in range(self.training_iters):
            
            gradient = np.zeros(self.w_hat.shape) 
            
            
            for point, yi in zip(X, y_train):
                gradient +=  (point * self.w_hat - yi) * point
            
            gradient *= self.learning_rate 

            self.w_hat -= gradient
            
            l = ((y_train - self.predict(X_train)) ** 2).sum() 
                        
            loss.append(l) 
            if self.show_learning:
                
                self._show_state(X_train, y_train, loss) 
            
        
    def predict(self, X_test):
        
        if len(X_test.values.shape) < 2:
            X = X_test.values.reshape(-1,1)
        X = np.insert(X, 0, 1, 1)
        
        return np.dot(X, self.w_hat) 
    
    
    def _show_state(self, X_train, y_train, loss):
      
        lb = L(y_train, self.w_hat[0] + 3 * X_train) 
        lw = L(y_train, 5 + self.w_hat[1] * X_train) 

        
        plt.subplot(221)
        plt.scatter(X_train, y_train, s= 10)
        plt.plot(X_train, self.predict(X_train), c='r')
        plt.title('$y = b + w x$')
        plt.tick_params(labelsize=9, labelleft=False, labelbottom = False)
        plt.grid(True)

        
        plt.subplot(222)
        plt.plot(range(len(loss)), loss)
        plt.title('Custo')
        plt.tick_params(labelsize=9, labelleft=False, labelbottom = False)
        plt.grid(True)

        
        plt.subplot(223)
        plt.plot( np.linspace(-10,20,20), self.b_loss)
        plt.scatter(self.w_hat[0], lb, c = 'k')
        plt.title('Custo em $\hat{b}$')
        plt.tick_params(labelleft=False)
        plt.grid(True)
        
        
        plt.subplot(224)
        plt.plot(np.linspace(0,6,20), self.w1_loss)
        plt.scatter(self.w_hat[1], lw, c = 'k')
        plt.title('Custo em $\hat{w}$')
        plt.grid(True)
        plt.tick_params(labelleft=False)
        
        plt.tight_layout()
        display.display(plt.gcf())
        display.clear_output(wait=True)
        plt.clf() 

regr = linear_regr(learning_rate=0.0005, training_iters=30, show_learning=True)
regr.fit(dados['x'], dados['y'])
```
